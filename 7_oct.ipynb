```python
#1
class bankaccount:
    def __init__(self,account_number,account_name,balance):
        self.account_number=account_number
        self.account_name=account_name
        self.balance=balance
    def deposit(self,amount):
        if amount>0:
            self.balance +=amount
            print("the amount of ",amount,"is been deposited")
            print("the new balance is ",self.balance)
        else:
            print("Invalid deposit amount")
    def withdraw(self,amount):
        if amount>0 and self.balance>=amount:
            self.balance -= amount
            print("the amount of ",amount,"is been credited ")
            print("the new balance is :",self.balance)
        else:
            print("Invalid amount")
a=bankaccount(214,"vaibhav",2100)
a.deposit(400)
a.withdraw(500)
```

```python
#3
class Vehicle:
    def __init__(self,plate_no,make,model,is_available=True):
        self.plate_no=plate_no
        self.model=model
        self.make=make
        self.is_available=is_available

class RentalSystem:
    def __init__(self):
        self.vehicles=[]
    
    def add_vehicle(self,vehicle):
        self.vehicles.append(vehicle)
        
    def rent_vehicle(self,plate_no):
        for vehicle in self.vehicles:
            if vehicle.plate_no == plate_no and vehicle.is_available:
                print(f"Vehicle is now renter {plate_no}")
                vehicle.is_available=False
                return
        print(f"Vehicle {plate_no} is not available")
    
    def return_vehicle(self,plate_no):
        for vehicle in self.vehicles:
            if vehicle.plate_no == plate_no and not vehicle.is_available:
                print(f"Vehicle is return {plate_no}")
                vehicle.is_available = True
                return
        print(f"Vehicle {plate_no} is not return")
    
    def display_available_vehicles(self):
        available_vehicles = [vehicle for vehicle in self.vehicles if vehicle.is_available]
        if available_vehicles:
            print("Available vehicles:")
            for vehicle in available_vehicles:
                print(f"{vehicle.make} {vehicle.model} - Plate Number: {vehicle.plate_no}")
        else:
            print("No vehicles available.")

system = RentalSystem()
vehicle1 = Vehicle("ABC123", "Toyota", "Camry")
vehicle2 = Vehicle("XYZ456", "Honda", "Civic")
system.add_vehicle(vehicle1)
system.add_vehicle(vehicle2)

system.display_available_vehicles()

system.rent_vehicle("ABC123")
system.rent_vehicle("XYZ456")
system.return_vehicle("XYZ456")

system.display_available_vehicles()
```

```python
#4
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"Borrowed book: {self.title} by {self.author}")
        else:
            print(f"Book '{self.title}' is already borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"Returned book: {self.title} by {self.author}")
        else:
            print(f"Book '{self.title}' is already available.")


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_available_books(self):
        print("Available Books:")
        for book in self.books:
            if not book.is_borrowed:
                print(f"- {book.title} by {book.author}")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and not book.is_borrowed:
                book.borrow()
                return
        print(f"Book '{title}' is not found or already borrowed.")

    def return_book(self, title):
        for book in self.books:
            if book.title == title and book.is_borrowed:
                book.return_book()
                return
        print(f"Book '{title}' is not found or already available.")


# Example usage
library = Library()

library.add_book(Book("The Lord of the Rings", "J.R.R. Tolkien", "9780261102694"))
library.add_book(Book("Pride and Prejudice", "Jane Austen", "9780140435221"))
library.add_book(Book("To Kill a Mockingbird", "Harper Lee", "9780446310727"))

library.display_available_books()

library.borrow_book("Pride and Prejudice")
library.borrow_book("The Catcher in the Rye") 
library.display_available_books()

library.return_book("Pride and Prejudice")
library.return_book("The Lord of the Rings") 

library.display_available_books()
```

```python
#5
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"ID: {self.product_id}, Name: {self.name}, Price: ${self.price:.2f}, Quantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = {} 

    def add_product(self, product):
        self.products[product.product_id] = product

    def update_product_quantity(self, product_id, change):
        if product_id in self.products:
            self.products[product_id].quantity += change
            if self.products[product_id].quantity < 0:
                print(f"Error: Cannot update quantity to a negative value for product '{self.products[product_id].name}'.")
                self.products[product_id].quantity -= change 
            else:
                print(f"Updated quantity of '{self.products[product_id].name}' to {self.products[product_id].quantity}.")
        else:
            print(f"Product with ID {product_id} not found in inventory.")

    def display_available_products(self):
        print("Available Products:")
        for product in self.products.values():
            if product.quantity > 0:
                print(product)


# Example usage
inventory = Inventory()

product1 = Product(1001, "T-Shirt", 19.99, 10)
product2 = Product(1002, "Mug", 9.99, 5)

inventory.add_product(product1)
inventory.add_product(product2)

inventory.display_available_products()

inventory.update_product_quantity(1001, 5)  
inventory.update_product_quantity(1003, -2)  
inventory.update_product_quantity(1002, -10)  

inventory.display_available_products()
```

```python
#6
class Shape:
    def __init__(self):
        pass

    def calculate_area(self):
        raise NotImplementedError("Subclasses must implement their own calculate_area method")

    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses must implement their own calculate_perimeter method")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        return self.side_length ** 2

    def calculate_perimeter(self):
        return 4 * self.side_length


# Example usage
rectangle = Rectangle(5, 3)
square = Square(4)

print(f"Rectangle area: {rectangle.calculate_area()}")
print(f"Rectangle perimeter: {rectangle.calculate_perimeter()}")

print(f"Square area: {square.calculate_area()}")
print(f"Square perimeter: {square.calculate_perimeter()}")
```

```python
#7
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades  

    def calculate_average_grade(self):
        """Calculates the average grade of the student.

        Returns:
            float: The average grade, or None if there are no grades.
        """
        if not self.grades:
            return None  
        return sum(self.grades) / len(self.grades)

    def display_details(self):
        """Prints the student's details, including ID, name, and average grade."""
        average_grade = self.calculate_average_grade()
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if average_grade is not None:
            print(f"Average Grade: {average_grade:.2f}")
        else:
            print("No grades available.")


student1 = Student(12345, "John Doe", [85, 90, 78])
student2 = Student(54321, "Jane Smith", []) 

student1.display_details()
student2.display_details()

```

```python
#9
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []  

    def add_post(self, content):
        self.posts.append(content)
        print(f"Post added successfully!")

    def display_posts(self):
        if not self.posts:
            print("No posts available yet.")
            return

        print(f"\nPosts for {self.username}:")
        for index, post in enumerate(self.posts):
            print(f"Post {index + 1}:")
            print(post)

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if not matching_posts:
            print(f"No posts found containing the keyword '{keyword}'.")
            return []

        print(f"\nSearch results for '{keyword}':")
        for post in matching_posts:
            print(post)

        return matching_posts

profile = SocialMediaProfile("johndoe")

profile.add_post("This is my first post!")
profile.add_post("Hello, everyone! How are you today?")
profile.add_post("Just finished reading a great book. Highly recommend it!")

profile.display_posts()

profile.search_posts("book")
profile.search_posts("travel")  
```

```python
#10
class ToDoList:
    def __init__(self):
        self.tasks = []  # List to store tasks as dictionaries

    def add_task(self, task, due_date):
    
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})
        print(f"Task added successfully!")

    def mark_task_completed(self, task_description):
    
        for task in self.tasks:
            if task["task"] == task_description:
                task["completed"] = True
                print(f"Task '{task_description}' marked as completed.")
                return True
        print(f"Task '{task_description}' not found in the list.")
        return False

    def display_pending_tasks(self):
       
        if not self.tasks:
            print("No tasks in the list yet.")
            return

        print("\nPending Tasks:")
        for index, task in enumerate(self.tasks, start=1):
            if not task["completed"]:
                due_date_str = f" (Due: {task['due_date']})" if task["due_date"] else ""
                print(f"{index}. {task['task']}{due_date_str}")


todo_list = ToDoList()

todo_list.add_task("Buy groceries", "Tomorrow")
todo_list.add_task("Finish report", "")
todo_list.add_task("Clean the room", "Today")

todo_list.display_pending_tasks()

todo_list.mark_task_completed("Buy groceries")
todo_list.mark_task_completed("Wash car") 

todo_list.display_pending_tasks()
```
